#### Scatter plots
ใช้สำหรับดูการกระจายของข้อมูล

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,21)
y = np.random.randint(1,100,20)

plt.scatter(x,y,marker = 'o', c='r',s=100,ec='b',linewidths=2)
#plt.grid()
plt.show()
```

#### Bar charts

```python
x = ['Mon','Tue','Wed','Thu','Fri']
y = [50,70,35,90,40]

plt.bar(x,y,color = 'r',alpha=0.5,hatch = '\/-|',lw=3 ,ec='k')
# add new other plot
plt.plot(x,y,c='yellow' , lw= 5)
plt.show()
```

```python
plt.barh(x,y,color = 'b',alpha=0.5,hatch = '+')
plt.show()
```

#### Histogram
ใช้กับข้อมูลที่เป็นตัวเลขเยอะๆ

```python
data = np.random.randint(1,100,200)
plt.hist(data,bins=20,ec='k')
plt.show()
```

#### Pie charts

```python
data = [45,20,25,5,10]
lb = ['Chrome','Edge','Firefox','Opera','Safari']
plt.pie(data,labels=lb,explode=[0,0,0,0,0],shadow=True,autopct='%.1f%%')
plt.show()
```

#### Box plots ( ใช้ของ seaborn ดีกว่า )

```python
dic = {
    'Math' : [10,40,78,55,60],
    'Eng' : [99,23,56,87,65],
    'Thai' : [78,50,60,34,25],
}

df = pd.DataFrame(dic)
data = [df['Math'],df['Eng'],df['Thai']]

plt.boxplot(data,vert=True,widths=0.3,labels=['Math','Eng','Thai'],patch_artist=True)
plt.show()
```

#### 3D plots

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.random.randint(1, 100, 5)
y = np.random.randint(1, 100, 5)
z = np.random.randint(1, 100, 5)

ax.plot3D(x, y, z, 'o-.r')
plt.show()
```
