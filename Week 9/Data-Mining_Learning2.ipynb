```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('mpg')
```

```python
df.info()
```

```python
df.head()
```

```python
sns.scatterplot(data=df , x='displacement',y='horsepower')
plt.show()
```

```python
df2 = df[['displacement','horsepower']].dropna()
```

```python
df2.head()
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=4,random_state=0)
x = df2['displacement']
y = df2['horsepower']
model.fit(df2)
```

```python
model.cluster_centers_
```

```python
sns.scatterplot(data=df2 , x='displacement',y='horsepower')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],color='#000')
plt.show()
```

```python
model.labels_
```

```python
plt.figure(figsize=[3,2])
sns.scatterplot(data=df2 , x='displacement',y='horsepower',hue=model.labels_ , palette='Set1')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],color='#000')
plt.show()
```

```python
df3 = df[['acceleration','weight']].dropna()
```

```python
model = KMeans(n_clusters=4,random_state=0)
model.fit(df3)

plt.figure(figsize=[4,3])
sns.scatterplot(data=df3 , x='acceleration',y='weight',hue=model.labels_ , palette='Set1')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],color='#000')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df3)
```

```python
df3[['acceleration','weight']] = scaler.transform(df3)
df4= df[['acceleration','weight']].dropna()
```

```python
df3
```

```python
df4
```

```python
model = KMeans(n_clusters=3,random_state=0)
model.fit(df3)

plt.figure(figsize=[4,3])
sns.scatterplot(data=df3 , x='acceleration',y='weight',hue=model.labels_ , palette='Set1')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],color='#000')
plt.show()
```

```python
df4[['acceleration','weight']] = scaler.inverse_transform(df3)

plt.figure(figsize=[4,3])
sns.scatterplot(data=df4 , x='acceleration',y='weight',hue=model.labels_ , palette='Set1')
center = scaler.inverse_transform(model.cluster_centers_)
plt.scatter(center[:,0],center[:,1],color='#000',marker='*')
plt.show()
```
