# Matplotlib

### import

```python
import numpy  as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dic = {"Day":['Sun','Mon','Tue','Wed','Thu','Fri','Sat'],
       "Temp":[24,25,29,35,23,20,23]}
df = pd.DataFrame(dic)
x = df['Day']
y = df['Temp']
plt.plot(x,y)
plt.show()
```

#### plt.plot(ข้อมูลแกน x , ข้อมูลแกน y , typeเส้น(ls),ความหนา(lw)marker = 'กำหนดตัวมาร์ค')

```python
plt.plot(df['Day'],df['Temp'], ls='--',lw='10')
```

```python
plt.plot(x, y, color=(1,0.5,0))
```

```python
plt.plot(df['Day'],df['Temp'], ls='-',lw='5',marker='o',ms=10,mec='red',mfc='red')
```

```python
plt.plot(df['Day'],df['Temp'], "o-r",lw='3',mfc="black",ms=10)
```

#### plt.title , plt.xlabel and plt.xlabel

```python
plt.plot(df['Day'],df['Temp'], "o-r",lw='3',mfc="black",ms=10,color='#773590')
plt.title('Temp')
plt.xlabel('Day')
plt.ylabel('Temp')
plt.xticks(rotation = 90)
plt.yticks(np.arange(1,50,5))
plt.show()
```

```python
dic = {"Day":['Sun','Mon','Tue','Wed','Thu','Fri','Sat'],
       "Temp1":[24,25,29,35,23,20,23],
       "Temp2":[10,14,9,12,11,15,17]}
df = pd.DataFrame(dic)
x = df['Day']
y = df['Temp1']
z = df['Temp2']
plt.plot(x,y,'s-.r')
plt.plot(x,z,'o:g')
plt.show()
```

#### Matplotlib - legend

```python
plt.plot(x,y,'s-.r',label='T1')
plt.plot(x,z,'o:g',label='T2')
plt.legend(loc='best')
```

```python
fig , axes = plt.subplots(nrows=2 , ncols=2)
axes[0,0].plot(x,y,'o-.r')
axes[0,0].set_title("Week 1 ")
axes[0,1].plot(x,z,'o-g')
axes[0,1].set_title("Week 2 ")
axes[1,0].plot(y,x,'o:b')
axes[1,0].set_title("Week 3 ")
axes[1,1].plot(y,z,'o--w')
axes[1,1].set_title("Week 4 ")
fig.tight_layout()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(df['Day'],df['Temp1'],'s--r')
axes1.set_title("Week 1")
# axes1.set_xticks('Day')
axes2 = fig.add_axes([1.2,0,1,1])
axes2.plot(df['Day'],df['Temp2'],'s--g')
axes2.set_title("Week 2")
fig.tight_layout()
```
