## Data

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x = np.arange(0,100)
y = x*2
z = x**2
df = pd.read_csv('Superstore.csv',encoding = 'iso-8859-1')
```

## Exercise 1


```python
plt.plot(x,y)
plt.title("Title")
plt.xlabel('x')
plt.ylabel('y')
```

## Exercise 2

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,y)
axes1.set_title("Title")
axes1.set_xlabel('x')
axes1.set_ylabel('y')


axes2 = fig.add_axes([0.2,0.6,0.3,0.3])
axes2.plot(x,y)
axes2.set_title("Title")
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_yticks(np.arange(0,300,100))

plt.show()
```

## Exercise 3



**ใช้  arrays x, y และ z เพื่อทำการ plot บนแกนที่สร้างจากข้อที่แล้ว  (Notice อย่าลืมกำหนด  x - limits และ y - limits )**

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,z)
axes1.set_title("Title")
axes1.set_xlabel('x')
axes1.set_ylabel('z')


axes2 = fig.add_axes([0.2,0.6,0.3,0.3])
axes2.plot(x,y)
axes2.set_title("Zoom")
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_ylim(30,50)
axes2.set_xlim(20.0,22.0)
```

## Exercise 4

**จงใช้คำสั่ง plt.subplots(nrows=1, ncols=2)**

**จากนั้นให้ทำการ plot (x,y) และ plot (x,z) บนแกน axes  และให้ใช้งานคำสั่ง linewidth and style เพื่อตกแต่งเส้นของกราฟ**

```python
# plt.style.use('default')
plt.style.use('Solarize_Light2')
fig , axes = plt.subplots(nrows=1 , ncols=2 ,figsize=[8,3])
axes[0].plot(x, y ,'--r',lw=3)
axes[0].set_xticks(np.arange(0,120,20))
axes[0].set_yticks(np.arange(0,250,50))
axes[1].plot(x, z , '-g',lw=3)
axes[1].set_xticks(np.arange(0,120,20))
fig.tight_layout()
plt.show()
```

#### โค้ดต่อไปนี้ใช้ในสองข้อสุดท้าย

```python
df['Category'].unique()
```

```python
df[ df['Category'] == 'Furniture' ].groupby('Year').sum()['Profit']
```

##### Dictionary of Category

```python
arr_df = {}
for i in range(0,df['Category'].nunique()) :
    arr_df[df['Category'].unique()[i]] = df[ df['Category'] == df['Category'].unique()[i] ].groupby('Year').sum()['Profit']
```

```python
x = arr_df['Furniture'].index
y_Furniture = arr_df['Furniture']
y_Office = arr_df['Office Supplies']
y_Tech = arr_df['Technology']
```

## จงแสดงกราฟรายได้ของแต่ละ Category ในแต่ละปีในกราฟเดียว 

```python
plt.plot(x,y_Furniture,'^-',label='Furniture')
plt.plot(x,y_Office,'o-',label='Office Supplies')
plt.plot(x,y_Tech,'s-',label='Technology')
plt.yticks(np.arange(10000,60000,10000))
plt.xticks([2014,2015,2016,2017])
plt.legend(loc='best')
plt.show()
```

## จงแสดงกราฟรายได้ของแต่ละ Category ในแต่ละปี แบบแยกกราฟ

```python
fig = plt.figure()
axes1 = fig.add_axes([0.5,1.2,1,1])
axes1.plot(x,y_Furniture,'o:g',label = 'Technology')
axes1.set_title('Technology')

axes2 = fig.add_axes([0,0,1,1])
axes2.plot(x,y_Office,'^--b',label = 'Furniture')
axes2.set_title('Furniture')

axes3 = fig.add_axes([1.2,0,1,1])
axes3.plot(x,y_Tech,'s-.r',label = 'Office Supplies')
axes3.set_title('Office Supplies')

axes1.set_xticks([2014,2015,2016,2017])
axes2.set_xticks([2014,2015,2016,2017])
axes3.set_xticks([2014,2015,2016,2017])

plt.show()
```
