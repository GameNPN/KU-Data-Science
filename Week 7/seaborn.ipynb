```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head
```

```python
plt.figure(figsize=[4,2])
sns.histplot(tips['tip'],bins=20,kde = True)
plt.show()
```

```python
plt.figure(figsize=[4,2])
sns.histplot(tips['total_bill'],bins=20,kde = True)
plt.show()
```

```python
plt.figure(figsize=[3,2])
sns.countplot(data=tips,x='time')
plt.show()
```

```python
plt.figure(figsize=[3,2])
sns.barplot(data=tips , x = 'sex' , y = 'tip')
plt.show()
```

```python
plt.figure(figsize=[3,2])
sns.countplot(data=tips , x = 'sex')
plt.show()
```

```python
plt.figure(figsize=[3,2])
sns.histplot(data=tips , x = 'sex' palette='rocket')
plt.show()
```
