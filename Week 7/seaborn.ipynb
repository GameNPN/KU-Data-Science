```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head
```

```python
plt.figure(figsize=[4,2])
sns.histplot(tips['tip'],bins=20,kde = True)
plt.show()
```

```python
plt.figure(figsize=[4,2])
sns.histplot(tips['total_bill'],bins=20,kde = True)
plt.show()
```

```python
plt.figure(figsize=[3,2])
sns.countplot(data=tips,x='time')
plt.show()
```

```python
plt.figure(figsize=[3,2])
sns.barplot(data=tips , x = 'sex' , y = 'tip')
plt.show()
```

```python
plt.figure(figsize=[3,2])
sns.countplot(data=tips , x = 'sex')
plt.show()
```

```python
plt.figure(figsize=[3,2])
sns.histplot(data=tips , x = 'sex',palette='magma')
plt.show()
```

```python
plt.figure(figsize=[4,2])
sns.boxplot(data= tips , x='day',y='tip')
plt.show()
```

```python
plt.figure(figsize=[6,4])
sns.boxplot(data= tips , x='day',y='tip' ,hue='sex')
plt.show()
```

```python
plt.figure(figsize=[8,5])
sns.violinplot(data= tips , x='day',y='tip')
plt.show()
```

```python
plt.figure(figsize=[8,5])
sns.violinplot(data= tips , x='day',y='tip',hue='sex',split='True')
plt.show()
```

```python
plt.figure(figsize=[6,4])
sns.boxplot(data= tips , x='day',y='tip' ,hue='sex')
sns.stripplot(data= tips , x='day',y='tip' ,hue='sex')
plt.show()
```

```python
plt.figure(figsize=[6,4])
sns.boxplot(data= tips , x='day',y='tip' ,hue='sex')
sns.stripplot(data= tips , x='day',y='tip' ,hue='sex',jitter=0)
plt.show()
```

```python
plt.figure(figsize=[6,4])
sns.stripplot(data= tips , x='day',y='tip' ,hue='sex',jitter=0)
plt.show()
```

```python
sns.jointplot(data=tips , x='total_bill',y='tip',kind='reg')
plt.show()
```

```python
sns.jointplot(data=tips , x='total_bill',y='tip',kind='hex')
```

```python
sns.pairplot(data=tips,hue='sex')
plt.show()
```

```python
g = sns.JointGrid(data=tips,x='total_bill',y='tip')
g.plot(sns.scatterplot,sns.boxplot)
plt.show()
```

```python
g = sns.PairGrid(data=tips,hue='sex')
g.map_diag(sns.jointplot)
g.map_lower(sns.boxplot)
g.map_upper()
plt.show()
```
